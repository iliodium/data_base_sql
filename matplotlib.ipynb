```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
ax.plot(t, s, color='blue', lw=2)
ln,=ax.plot(t+1, s+1, color='blue', lw=2)
```
